```python
import pandas as pd
import matplotlib.pyplot as plt

# Load the dataset
data = pd.read_csv('customers.csv')

# Plotting with plt.plot as a line chart
plt.plot(data['Age'], data['Income'], alpha=0.5)
plt.title('Age vs. Income')
plt.xlabel('Age')
plt.ylabel('Income')
plt.grid(True)
plt.show()
```

```python
# Bin the data by age
bins = pd.cut(data['Age'], bins=20)
binned_data = data.groupby(bins)['Income'].mean()

# Plotting with plt.plot as a line chart
plt.plot(binned_data.index.astype(str), binned_data.values, marker='o')
plt.title('Average Income by Age Group')
plt.xlabel('Age Group')
plt.ylabel('Average Income')
plt.xticks(rotation=45)  # Rotate x-axis labels for better readability
plt.grid(True)
plt.show()
```

```python
# Downsample the data
sampled_data = data.sample(frac=0.01, random_state=42)  # Select 1% of the data randomly

# Plotting with plt.plot as a line chart
plt.plot(sampled_data['Age'], sampled_data['Income'], 'bo', alpha=0.5)
plt.title('Age vs. Income (Downsampled)')
plt.xlabel('Age')
plt.ylabel('Income')
plt.grid(True)
plt.show()
```

```python
# Creating a new figure with plt.figure
plt.figure(figsize=(10, 6))
plt.hist(data['Socks_Owned'], bins=20, color='skyblue', edgecolor='black')
plt.title('Distribution of Socks Owned')
plt.xlabel('Number of Socks Owned')
plt.ylabel('Frequency')
plt.grid(True)
plt.show()
```

```python
# Creating subplots with plt.subplot
fig, axs = plt.subplots(3, 1, figsize=(10, 8))
fig.suptitle('Age, Income and Location Distributions')
axs[0].hist(data['Age'], bins=30, color='orange', edgecolor='black')
axs[0].set_title('Distribution of Age')
axs[0].set_xlabel('Age')
axs[0].set_ylabel('Frequency')
axs[0].grid(True)

axs[1].hist(data['Income'], bins=30, color='green', edgecolor='black')
axs[1].set_title('Distribution of Income')
axs[1].set_xlabel('Income')
axs[1].set_ylabel('Frequency')
axs[1].grid(True)

# Plotting Number of Customers by City Location using bar chart
city_counts = data['City_Location'].value_counts()
axs[2].bar(city_counts.index, city_counts.values)
axs[2].set_title('Number of Customers by City Location')  # Use set_title instead of title
axs[2].set_xlabel('City Location')  # Use set_xlabel instead of xlabel
axs[2].set_ylabel('Number of Customers')  # Use set_ylabel instead of ylabel
axs[2].tick_params(axis='x', rotation=45)  # Use tick_params for rotation

plt.tight_layout()
plt.show()
```

```python
# Saving the figure with plt.savefig
plt.figure(figsize=(8, 6))
plt.hist(data['Age'], bins=30, color='purple', edgecolor='black')
plt.title('Distribution of Age')
plt.xlabel('Age')
plt.ylabel('Frequency')
plt.grid(True)
plt.savefig('age_distribution.png')
plt.show()
```

```python
import seaborn as sns

# Load the dataset
data = pd.read_csv('customers.csv')
# Set Seaborn style
sns.set_theme(style="whitegrid")

# Histogram
plt.figure(figsize=(10, 6))
sns.histplot(data['Age'], kde=False, bins=20, color='skyblue')
plt.title('Age Distribution')
plt.xlabel('Age')
plt.ylabel('Frequency')
plt.show()
```

```python
# KDE (Kernel Density Estimate)
plt.figure(figsize=(10, 6))
sns.kdeplot(data['Income'], fill=True, color='salmon')
plt.title('Kernel Density Estimate of Income')
plt.xlabel('Income')
plt.ylabel('Density')
plt.show()
```

```python
# Boxplot
plt.figure(figsize=(10, 6))
sns.boxplot(x='Gender', y='Income', data=data, palette='Set3')
plt.title('Income Distribution by Gender')
plt.xlabel('Gender')
plt.ylabel('Income')
plt.show()
```

```python
# Violin plot
plt.figure(figsize=(10, 6))
sns.violinplot(x='Gender', y='Income', data=data, palette='Set3')
plt.title('Income Distribution by Gender')
plt.xlabel('Gender')
plt.ylabel('Income')
plt.show()
```

```python
# Pair plot
sns.pairplot(data[['Age', 'Income', 'Socks_Owned', 'Frequent_Losers']], diag_kind='kde', markers='+')
plt.suptitle('Pair Plot of Age, Income, Socks Owned, and Frequent Losers', y=1.02)
plt.show()
```

```python
# Heatmap
plt.figure(figsize=(10, 6))
sns.heatmap(data[['Age', 'Income', 'Socks_Owned', 'Frequent_Losers']].corr(), annot=True, cmap='coolwarm')
plt.title('Correlation Heatmap')
plt.show()
```
