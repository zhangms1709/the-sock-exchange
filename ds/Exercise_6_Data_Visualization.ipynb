## Creating Basic Plots with Matplotlib
Matplotlib is a core Python library for data visualization. It allows you to create line plots, bar charts, scatter plots, and more.

**Why use Matplotlib?**
- Detailed control over plot appearance.
- Useful for exploring and communicating data insights.

**Use Case:**
Create a line plot showing the trend of values over time or categories.


```python
import matplotlib.pyplot as plt

# Sample data
months = ['Jan', 'Feb', 'Mar', 'Apr', 'May']
sales = [250, 270, 300, 310, 400]

# Create a line plot
plt.plot(months, sales)
plt.title('Monthly Sales')
plt.xlabel('Month')
plt.ylabel('Sales ($)')
plt.grid(True)
plt.show()
```

## Comparing Values: Bar Charts

**Bar Charts:**

Bar charts are useful for comparing quantities across categories. They use rectangular bars with lengths proportional to the values they represent.

- **What it shows:** The comparison of values across discrete categories.
- **When it is used:** To compare sales per product, votes per candidate, counts per category, or any other data where you need to compare values across distinct groups. Bar charts can be vertical (plt.bar()) or horizontal (plt.barh()).

```python
import matplotlib.pyplot as plt

categories = ['Apples', 'Bananas', 'Cherries', 'Dates']
values = [25, 40, 30, 10]

plt.bar(categories, values)
plt.title('Fruit Sales')
plt.xlabel('Fruit')
plt.ylabel('Units Sold')
plt.show()
```

## Comparing Values: Pie Charts

**Pie Charts:**

Pie charts are great for visualizing part-to-whole relationships. They represent data as slices of a circle, where the size of each slice is proportional to the percentage it represents of the whole.

- **What it shows:** The proportion or percentage of different categories within a whole dataset.
- **When it is used:** To show how a total is divided into different parts, such as budget allocation, market share, or survey responses. They are most effective when you have a small number of categories.

```python
labels = ['R&D', 'Marketing', 'Sales', 'HR']
sizes = [30, 25, 35, 10]

plt.pie(sizes, labels=labels, autopct='%1.1f%%', startangle=90)
plt.axis('equal')
plt.title('Department Budget Allocation')
plt.show()
```

## Showing Relationships: Scatter Plots

**Scatter Plots:**

Scatter plots are used to display the relationship between two numerical variables. Each point on the scatter plot represents a single data point, with its position determined by its values for the two variables.

- **What it shows:** The nature and strength of the relationship between two variables (e.g., positive correlation, negative correlation, no correlation), the spread of data points, and potential outliers.
- **When it is used:** To explore potential relationships or correlations between two numerical variables, identify clusters of data points, and visualize the distribution of data in two dimensions.

```python
import numpy as np

x = np.random.rand(50)
y = 3 * x + np.random.normal(0, 0.2, 50)

plt.scatter(x, y)
plt.title('Hours Studied vs Exam Score')
plt.xlabel('Hours Studied')
plt.ylabel('Exam Score')
plt.show()
```

## Showing Relationships: Line Plots

**Line Plots:**

Line plots are used to display the relationship between two numerical variables, typically showing how one variable changes as the other increases. They are particularly effective for visualizing trends over a continuous interval or time.

- **What it shows:** Trends, patterns, and changes in data over a continuous range (like time, distance, or another ordered sequence).
- **When it is used:** To show how a quantity changes over time (e.g., stock prices, temperature), to compare trends of multiple variables, or to display the relationship between two numerical variables where the x-axis represents a continuous progression.

```python
x = np.linspace(0, 12, 100)
y = np.sin(x)

plt.plot(x, y)
plt.title('Seasonal Pattern')
plt.xlabel('Time')
plt.ylabel('Value')
plt.grid(True)
plt.show()
```

## Showing Trends Over Time: Heatmaps

**Heatmaps:**

Heatmaps are visualizations that use color intensity to represent the magnitude of values in a matrix or table. They are particularly useful for showing patterns, variations, and correlations in data.

- **What it shows:** The strength of relationships between two variables (in a correlation heatmap), or the variation of a single variable across two dimensions (like time and location).
- **When it is used:** To quickly identify areas of high or low values, visualize correlation matrices, show patterns in time-series data, or compare categories across multiple features.

```python
import seaborn as sns
import pandas as pd

data = np.random.rand(6, 6)
df = pd.DataFrame(data, columns=list('ABCDEF'))

sns.heatmap(df, annot=True, cmap='YlGnBu')
plt.title('Random Heatmap')
plt.show()
```

## Showing Distributions: Histograms & Box Plots

**Histograms:**

A histogram is a graphical representation of the distribution of a dataset. It groups data into bins and shows how many data points fall into each bin.

- **What it shows:** The shape of the distribution (e.g., normal, skewed), the central tendency (where most of the data is clustered), and the spread or variability of the data.
- **When it is used:** To understand the underlying distribution of a single numerical variable, identify outliers, and compare distributions of different datasets.

**Box Plots:**

A box plot (also known as a box and whisker plot) is a standardized way to display the distribution of data based on a five-number summary: minimum, first quartile (Q1), median, third quartile (Q3), and maximum.

- **What it shows:** The median (the line in the middle of the box), the interquartile range (IQR, the box itself, which contains the middle 50% of the data), the range of the data (whiskers), and potential outliers (points outside the whiskers).
- **When it is used:** To compare the distributions of one or more groups, identify potential outliers, and get a quick summary of the data's spread and central tendency. They are particularly useful for comparing distributions across different categories.

```python
# Histogram
data = np.random.randn(1000)

plt.hist(data, bins=30, color='teal')
plt.title('Histogram of Exam Scores')
plt.xlabel('Score')
plt.ylabel('Frequency')
plt.show()

# Box Plot
group_a = np.random.normal(70, 10, 100)
group_b = np.random.normal(60, 15, 100)

plt.boxplot([group_a, group_b], tick_labels=['Group A', 'Group B'])
plt.title('Score Distribution by Group')
plt.ylabel('Scores')
plt.show()
```
