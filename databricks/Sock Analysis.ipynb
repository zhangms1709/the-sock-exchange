```python
socksDF = spark.read.format("json").load("dbfs:/sock_data1111.json")
```

```python
import seaborn as sns
import matplotlib.pyplot as plt
```

```python
size_distribution = socksDF.groupBy("Size").count().toPandas()
sns.barplot(x="Size", y="count", data=size_distribution)
plt.title("Distribution of Socks by Size")
plt.show()
```

```python
size_distribution = socksDF.groupBy("Material").count().toPandas()
sns.barplot(x="Material", y="count", data=size_distribution)
plt.title("Distribution of Socks by Material")
plt.show()
```

```python
size_distribution = socksDF.groupBy("Pattern").count().toPandas()
sns.barplot(x="Pattern", y="count", data=size_distribution)
plt.title("Distribution of Socks by Pattern")
plt.show()
```

```python
sns.displot(data=socksDF.toPandas(), x="Pattern", hue="Material", row="Size", multiple="stack")
plt.title("Distribution of Socks by Pattern and Material")
plt.show()
```
